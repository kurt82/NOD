```python
# zastosuj metodę groupby oddzielnie do każdej części, a następnie połącz wynik w nową ramkę danych
new_df = pd.DataFrame() # pusta ramka danych
for chunk_df in pd.read_csv('IHME_PREM_WMN_HEALTH_2020_Y2011M10D11.csv', 
                      chunksize = 50000):
    result = chunk_df.groupby(['geography', 'age']).agg(
        country_poland=('country', lambda x: (x == 'Poland').sum()),
        religion_count=('religion', 'count')
    )
    new_df = pd.concat([new_df,result])
```
